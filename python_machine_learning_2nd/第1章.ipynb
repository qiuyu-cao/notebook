# 第一章 什么是深度学习

## 1.1 人工智能、机器学习和深度学习

### 1.1.1 人工智能

- 人工智能诞生于 20 世纪 50 年代
- 人工智能可以被描述为试图将通常由人类完成的智力任务自动化

### 1.1.2 机器学习

- 人工智能先驱阿兰 • 图灵(Alan Turing)
> **图灵测试**的核心思想是通过对话来判断机器是否具有人类水平的智能。测试中，一个人类与一台机器进行对话，如果裁判员不能分辨对话双方哪个是人，哪个是机器，则视为机器通过了测试。

### 1.1.3 从数据中学习规则与表示

进行机器学习的3个要素：
- 输入数据
- 预期输出的示例
- 衡量算法效果的方法

### 1.1.4 深度学习之“深度”

- 深度学习中的“深度”表示的是一系列连续的表示层，层数被称为模型的深度。
- 这些分层表示是通过**神经网络**（neural network）的模型学习得到的。
- 其他机器学习通常仅学习一两层的数据表示，因此被称为浅层学习（shallow learning）。
- 神经网络的理解：类似与信息蒸馏（information distillation），想象一个Input是一个手写数字图片的像素点数据，这些数据穿过连续的过滤器，纯度越来越高，最终被识别为具体的数字（Output）。


### 1.1.5 用三种图理解深度学习的工作原理

- 每一层的权重（weight）保存了该层对输入数据所作的具体操作。
- 权重有时也被称为该层的参数（parameter）。
- 损失函数（loss function）用于衡量训练模型时，输出与预期结果之间的距离。又是也称为目标函数（objective function）或代价函数（cost function）
- 优化器（optimizer）根据损失值对模型的权重进行微调，它实现了所谓的**反向传播**（backpropagation）算法，是深度学习的核心算法。



- **训练循环**（training loop）是通过重复足够多的跌打训练，使得损失值讲到最小的过程。

## 1.2 深度学习之前：机器学习简史

深度学习并不一定总是解决问题的正确工具：有时没有足够多的数据，深度学习不适用。有时其他算法可以更好地解决问题。

### 1.2.1 概率建模

**概率建模**（probabilistic modeling）是统计学原理在数据分析中的应用。
- 贝叶斯算法是一类机器学习分类器，它假设输入数据的特征都是独立的。18世纪就已经开始用人工计算的方法进行应用，1950s 才在计算机上实现。
- logistic 回归（logistic regression，简称 logreg）

### 1.2.2 早期神经网络

- 反向传播算法的出现推动了神经网络的应用
- 神经网络的第一个成功案例来自 1989 年的贝尔实验室，当时 Yann LeCun 将卷积神经网络的早期思想与方向传播算法相结合，用于手写数字的分类问题。

### 1.2.3 核方法

- **核方法**（kernel method）是一组分析算法。其中最有名的就是**支持向量机**（support vector machine，SVM）.
- SVM 通过两步来寻找“决策边界”：
    - 将数据映射到新的高维表示，此时决策边界可以用一个超平面来表示；
    - 尽量让超平面与每个类别最近的数据点之间的距离最大化（间隔最大化，maximizing the mergin）
- SVM 是一种浅层方法，很难扩展到大型数据集。

### 1.2.4 决策树、随机森林和梯度提升机

- **决策树**（decision tree）是类似于流程图的结构。2021年，决策树往往比核方法更受欢迎。
- **随机森林**（random forest）：首先构建许多专门的决策树，然后将他们的输出集成在一起。适用于各种各样的问题，几乎总是第二好的算法。
- **梯度提升机**（grandient boosting machine）通过构建多个弱预测模型（通常是决策树），并将它们的结果组合起来以提高预测准确性。梯度提升机的核心思想是逐步添加模型，每个新模型都尝试纠正前一个模型的错误。**是目前处理非感知数据最好的算法。**

### 1.2.5 回到神经网络

- 卷积神经网络（convert）适用于所有感知任务。

### 1.2.6 深度学习有何不同

- 深度学习在很多问题上表现出更好的性能，并且使解决问题变得更加简单。
- 浅层学习往往需要对数据类型进行转化以适合于对应的方法（**特征工程**）
- 共同学习与贪婪学习

### 1.2.7 机器学习现状

- 一个了解现状的方法：Kaggle上的机器学习竞赛。
- 梯度提升树用于浅层学习问题（scikit-learn、XGBoost 或 LightGBM）
- 深度学习用于感知问题 （Keras 结合 TensorFlow）
